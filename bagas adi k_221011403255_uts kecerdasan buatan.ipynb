```python
# Data perusahaan
company_detail_list = [
    {'name': 'Company 1', 'domain': 'Retail', 'country': 'United States'},
    {'name': 'Company 2', 'domain': 'Technology', 'country': 'United Kingdom'},
    {'name': 'Company 3', 'domain': 'Healthcare', 'country': 'United States'}
]

# Data karyawan
employee_detail_list = [
    {'name': 'EMP-0001', 'full_name': 'John Doe', 'company': 'Company 1', 'nationality': 'Australia'},
    {'name': 'EMP-0002', 'full_name': 'Tom Smith', 'company': 'Company 2', 'nationality': 'United States'},
    {'name': 'EMP-0003', 'full_name': 'Andrew Sebastian', 'company': 'Company 3', 'nationality': 'United States'},
    {'name': 'EMP-0005', 'full_name': 'Ying Han Tan', 'company': 'Company 1', 'nationality': 'Australia'},
    {'name': 'EMP-0015', 'full_name': 'Kenneth Ng', 'company': 'Company 3', 'nationality': 'United States'},
    {'name': 'EMP-0018', 'full_name': 'Rubby Lee', 'company': 'Company 2', 'nationality': 'Hong Kong'},
    {'name': 'EMP-0017', 'full_name': 'Robert White', 'company': 'Company 1', 'nationality': 'United Kingdom'}
]
# Buat mapping antara nama perusahaan dan domainnya
company_domain_map = {company['name']: company['domain'] for company in company_detail_list}

# Buat dictionary untuk mengelompokkan karyawan berdasarkan domain
employees_by_domain = {}
# Loop untuk mengisi dictionary dengan karyawan sesuai domain
for employee in employee_detail_list:
    company = employee['company']
    domain = company_domain_map[company]  # Dapatkan domain perusahaan

    # Tambahkan karyawan ke dalam daftar di domain yang sesuai
    if domain not in employees_by_domain:
        employees_by_domain[domain] = []  # Buat list baru jika domain belum ada
    employees_by_domain[domain].append(employee['full_name'])

# Cetak hasil dalam format dictionary
print(employees_by_domain)
```

```python
# Data perusahaan
company_detail_list = [
    {'name': 'Company 1', 'domain': 'Retail', 'country': 'United States'},
    {'name': 'Company 2', 'domain': 'Technology', 'country': 'United Kingdom'},
    {'name': 'Company 3', 'domain': 'Healthcare', 'country': 'United States'}
]

# Data karyawan
employee_detail_list = [
    {'name': 'EMP-0001', 'full_name': 'John Doe', 'company': 'Company 1', 'nationality': 'Australia'},
    {'name': 'EMP-0002', 'full_name': 'Tom Smith', 'company': 'Company 2', 'nationality': 'United States'},
    {'name': 'EMP-0003', 'full_name': 'Andrew Sebastian', 'company': 'Company 3', 'nationality': 'United States'},
    {'name': 'EMP-0005', 'full_name': 'Ying Han Tan', 'company': 'Company 1', 'nationality': 'Australia'},
    {'name': 'EMP-0015', 'full_name': 'Kenneth Ng', 'company': 'Company 3', 'nationality': 'United States'},
    {'name': 'EMP-0018', 'full_name': 'Rubby Lee', 'company': 'Company 2', 'nationality': 'Hong Kong'},
    {'name': 'EMP-0017', 'full_name': 'Robert White', 'company': 'Company 1', 'nationality': 'United Kingdom'}
]
def get_employees_with_country():
    # Membuat peta antara nama perusahaan dan negara
    company_country_map = {company['name']: company['country'] for company in company_detail_list}
    
    # Membuat daftar karyawan dengan informasi negara
    employee_country_list = []
    
    for employee in employee_detail_list:
        company = employee['company']  # Ambil nama perusahaan karyawan
        country = company_country_map[company]  # Cari negara perusahaan
        
        # Buat dictionary untuk setiap karyawan
        employee_info = {
            'full_name': employee['full_name'],
            'company': company,
            'country': country
        }
        employee_country_list.append(employee_info)  # Tambahkan ke daftar
    
    return employee_country_list  # Kembalikan daftar karyawan

# Memanggil fungsi dan mencetak hasil dalam format yang diinginkan
result = get_employees_with_country()

# Cetak setiap karyawan dalam format JSON-like
for item in result:
    print("{")
    print(f'    "full_name": "{item["full_name"]}",')
    print(f'    "company": "{item["company"]}",')
    print(f'    "country": "{item["country"]}"')
    print("},")
```

```python
# Data karyawan
employee_detail_list = [
    {'name': 'EMP-0001', 'full_name': 'John Doe', 'company': 'Company 1', 'nationality': 'Australia'},
    {'name': 'EMP-0002', 'full_name': 'Tom Smith', 'company': 'Company 2', 'nationality': 'United States'},
    {'name': 'EMP-0003', 'full_name': 'Andrew Sebastian', 'company': 'Company 3', 'nationality': 'United States'},
    {'name': 'EMP-0005', 'full_name': 'Ying Han Tan', 'company': 'Company 1', 'nationality': 'Australia'},
    {'name': 'EMP-0015', 'full_name': 'Kenneth Ng', 'company': 'Company 3', 'nationality': 'United States'},
    {'name': 'EMP-0018', 'full_name': 'Rubby Lee', 'company': 'Company 2', 'nationality': 'Hong Kong'},
    {'name': 'EMP-0017', 'full_name': 'Robert White', 'company': 'Company 1', 'nationality': 'Hong Kong'}
]

def get_companies_with_employee_nationality():
    # Membuat dictionary untuk menghitung jumlah karyawan berdasarkan kebangsaan di setiap perusahaan
    company_nationality_count = {}

    # Looping semua karyawan untuk mengisi data kebangsaan per perusahaan
    for employee in employee_detail_list:
        company = employee['company']  # Ambil nama perusahaan
        nationality = employee['nationality']  # Ambil kebangsaan karyawan
    # Jika perusahaan belum ada di dictionary, tambahkan dengan dictionary kosong
        if company not in company_nationality_count:
            company_nationality_count[company] = {}
# Tambahkan atau update jumlah karyawan per kebangsaan
        if nationality in company_nationality_count[company]:
            company_nationality_count[company][nationality] += 1
        else:
            company_nationality_count[company][nationality] = 1

    # Konversi hasil ke dalam format list of dictionaries sesuai output yang diharapkan
    result = []
    for company, nationality_count in company_nationality_count.items():
        result.append({
            'company': company,
            'employee_nationality': nationality_count
        })

    return result

# Panggil fungsi dan cetak hasilnya
result = get_companies_with_employee_nationality()
# Cetak hasil dalam format rapi
for item in result:
    print("{")
    print(f'    "company": "{item["company"]}",')
    print(f'    "employee_nationality": {item["employee_nationality"]}')
    print("},")
```
